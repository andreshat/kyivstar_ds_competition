On the distribution plots shown above, there are a lot of skewed features. Applying Log Transformation to skewed distributed features could bring them closer to a normal distribution. That will help to conduct anomaly detection and perform more reliable predictions

```python
# function compares skewness before applying log and after, those features that improved their score will
# be returned as a list for post-processing 
def get_log_transform_features(df):
    non_log_transformed_skew_vals = df.skew(axis=0, skipna=True)
    
    # Drop categorical features
    df = df.select_dtypes(include='number')
        
    # log(x+1) to avoid infinite values if x = 0
    log_transformed_df = df.apply(lambda x: np.log(x+1), axis=1)
    
    log_transformed_skew_vals = log_transformed_df.skew(axis=0, skipna=True)
    
    # Catch Log(negative) that return NAN and replace with max
    log_transformed_skew_vals.fillna(non_log_transformed_skew_vals.max(), inplace=True)
    
    log_transform_feat_indexes = []
    skewed_feat_res = abs(non_log_transformed_skew_vals) - abs(log_transformed_skew_vals)
    feat_idx = 0
    for feat in skewed_feat_res:
        if feat > 0:
            log_transform_feat_indexes.append(feat_idx)
        feat_idx += 1
    
    log_transform_features = non_log_transformed_skew_vals.index[log_transform_feat_indexes]
    return log_transform_features
```

```python
# Exclude categorical one-hot encoding features
log_transform_features = get_log_transform_features(tabular_df.drop(exclude_cols, axis=1))
```

```python
log_tabular_df = tabular_df.copy()
log_tabular_df[log_transform_features] = log_tabular_df[log_transform_features].apply(lambda x: np.log(x+1), axis=1)
```

Identify outliers using 1.5 IQR rule

```python
# calculate lower and upper bounaries for every column to detect outliers
def get_outlier_boundaries(df):    
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

```python
# certain columns where lower and upper boundaries will be applied
outliers_columns = log_tabular_df.drop(exclude_cols, axis=1).columns
```

```python
lower_range, upper_range = get_outlier_boundaries(log_tabular_df.drop(exclude_cols, axis=1))
```

```python
# Replace outliers with NAN
df_out = log_tabular_df.copy()
df_out[outliers_columns] = df_out[outliers_columns][~((df_out[outliers_columns] < lower_range) | (df_out[outliers_columns] > upper_range))]
```

```python
def append_res_to_boxplot(results, df):
  i = 0
  while i < len(results[0]):
    line = []
    for num, ml in zip(results, mod_list): 
      line.append([num[i],ml])
    i = i+1
    df = df.append(pd.DataFrame(line, columns=['Score', 'ML']),ignore_index=True)
  return df
```
